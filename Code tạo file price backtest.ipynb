```python
os.chdir('../ami_price_data')

extension = 'csv'
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]

#combine all files in the list
ami_price = pd.concat([pd.read_csv(f) for f in all_filenames ])
ami_price['Date'] = pd.to_datetime(ami_price['Date/Time'],format='%m/%d/%Y %H:%M:%S')
ami_price['year'] = pd.DatetimeIndex(ami_price['Date/Time']).year
ami_price.reset_index(drop=True,inplace=True)
ami_price.to_csv( '..\\BCTC_data_done\\ami_pirce.csv', index=True, encoding='utf-8-sig')
ami_price
```
